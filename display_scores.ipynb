```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm
import numpy as np
```

```python
scores = pd.read_csv('scores/anime_scores.csv')
scores
```

```python
mal = scores[scores['source'] == 'MAL'].reset_index()
anidb = scores[scores['source'] == 'anidb'].reset_index()
```

```python
mal = mal.pivot(index='week', columns='name', values='score')
mal
```

```python
plt.figure(figsize=(10,7))
color = iter(cm.rainbow(np.linspace(0, 1, len(list(mal.columns)))))
for col in list(mal.columns):
    c = next(color)
    plt.plot(mal[col], label=col, color=c)
    
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
#plt.ylim(7.5, 9.0)
plt.show()
```

```python
anidb = anidb.pivot(index='week', columns='name', values='score')
anidb
```

```python
plt.figure(figsize=(10,7))
color = iter(cm.rainbow(np.linspace(0, 1, len(list(anidb.columns)))))
for col in list(anidb.columns):
    c = next(color)
    plt.plot(anidb[col], label=col, color=c)
    
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
#plt.ylim(6.5, 8.0)
plt.show()
```
